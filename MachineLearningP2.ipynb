```python
#def for fishers ratio

def rotation_m(deg):
  B = [(np.cos(deg),(-np.sin(deg))),(np.sin(deg),np.cos(deg))]
  return B

def mew_c(y_val):
  return (1/y_val.size)*np.sum(y_val)

def sigma_c(y_val,mew_val):
  return (1/y_val.size)*np.sum((y_val-mew_val)**2)

def fisher(mew1,mew2,sigma1,sigma2,size1,size2):
  num = (mew1-mew2)**2
  dem= (size1/(size1+size2))*sigma1 + (size2/(size1+size2))*sigma2
  return(num/dem)

```

```python
 # Fisher's ratio

na =1000
ma = (4, 6)
sa =[[4, 5], [5, 1]]
xa = np.random.multivariate_normal(ma, sa, na)

nb = 1000
mb =(10,12)
sb = [[6,3], [3, 4]]
xb =np.random.multivariate_normal(mb, sb, nb)
fishers = []
deg = []

 # Initial w

 wa = np.array([1.2,1.4])

 for i in range(0,361):
   deg.append(i)
   w_new = rotation_m(i*np.pi/180)   @ wa
   ya = w_new @ xa.transpose()
   yb =  w_new @ xb.transpose()

   mew_a = mew_c(ya)
   sigma_a = sigma_c(ya,mew_a)

   mew_b = mew_c(yb)
   sigma_b = sigma_c(yb,mew_b)
   val = fisher(mew_a,mew_b,sigma_a,sigma_b,na,nb)
   fishers.append(val)
plt.title("Graph comapring the fishers ratio comapred to angle of rotation of weight")
plt.xlabel("Degrees")
plt.ylabel("Fisher's Ratio")
plt.plot(deg,fishers)
print(np.argmax(fishers))
max_pos = np.argmax(fishers)
print(fishers[max_pos])

print(fishers[174])

  


```

LDA on multivariate data (Iris Dataset)

```python
#3 classes in iris dataset
# 4 features
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#class A -> y=0
#class b -> y=1
#class c -> y=2

Class_A =np.empty([0,4])
Class_B =np.empty([0,4])
Class_C = np.empty([0,4])
iter =0

print(Class_A)
for i in y:
  if i == 0:
    Class_A =np.append(Class_A,X[iter,:])

  elif i == 1:
    Class_B =np.append(Class_B,X[iter,:])
    
  elif i == 2:
    Class_C =np.append(Class_C,X[iter,:])
  iter+=1


Class_A = np.reshape(Class_A,(50,4))
Class_B =np.reshape(Class_B,(50,4))
Class_C =np.reshape(Class_C,(50,4))
```

```python

Mean_A =np.empty([0,4])
Mean_B = np.empty([0,4])
Mean_C = np.empty([0,4])
for i in range(4):
  F = (1/50)*np.sum(Class_A[:,i])
  Mean_A = np.append(Mean_A,F)

  F2 = (1/50)*np.sum(Class_B[:,i])
  Mean_B = np.append(Mean_B,F2)

  F3 = (1/50)*np.sum(Class_C[:,i])
  Mean_C= np.append(Mean_C,F3)


Total_Mean = 1/3*(Mean_A+Mean_B+Mean_C)
#print(Mean_A)
#print(Mean_B)
#print(Mean_C)
#print(Total_Mean)

m_vectors =np.empty([0,4])
m_vectors = np.stack(Mean_A)
m_vectors = np.append(m_vectors,Mean_B)
m_vectors = np.append(m_vectors,Mean_C)
m_vectors= np.reshape(m_vectors,(3,4))
print(m_vectors)

 
```

```python
Sigma_W = np.zeros((4,4))
for classes,m_vec in zip(range(0,3), m_vectors):
    mat_m = np.zeros((4,4))                  #  matrix for each class
    for row in X[y == classes]:
        #print(row,mv)
        row=row.reshape(4,1)
        m_vec =  m_vec.reshape(4,1) # make column vectors
        mat_m += (row-m_vec).dot((row-m_vec).T)
    Sigma_W += mat_m                             # sum  of each class matrices
print('Sigma_W Matrix:\n', Sigma_W)

```

```python

Sigma_B = np.zeros((4,4))
for mean in m_vectors: 
    mean = mean.reshape(4,1) # make column vector
    Total_Mean = Total_Mean.reshape(4,1) # make column vector
    Sigma_B +=  1/3*(mean - Total_Mean).dot((mean - Total_Mean).T)
print('Sigma_B Matrix:\n',Sigma_B)
```

```python
eigen_val, eigen_vec = np.linalg.eig(np.linalg.inv(Sigma_W).dot(Sigma_B))

for i in range(len(eigen_val)):
    eigvec = eigen_vec[:,i].reshape(4,1)   
    print( eigvec)
    print("Eigenvalue:",i, eigen_val[i])
```
